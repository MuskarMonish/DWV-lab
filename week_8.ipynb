```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("monthly_sales.csv")
```

```python
#profits vs expenses
fig=plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(df['profit'],df['expenses'])
ax1.set_title('profits vs expenses')
ax1.set_xlabel('profits')
ax1.set_ylabel('expenses')


ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(df['profit'], df['expenses'], color='r', marker='o')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('Profits')
ax2.set_ylabel('expenses')


ax3 = fig.add_subplot(2, 2, 3)
bar_labels = df['month']
bar_values = df['profit']
ax3.bar(bar_labels, bar_values, color='g', alpha=0.5)
ax3.set_title('Bar Plot')
ax3.set_xlabel('Profits')
ax3.set_ylabel('Expenses')
plt.xticks(df['month'],rotation=90)


ax4 = fig.add_subplot(2, 2, 4)
data = df['profit']
ax4.hist(data, bins=4, color='purple', alpha=0.7, edgecolor='black')
ax4.set_title('Histogram')
ax4.set_xlabel('Profits')
ax4.set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
#profits vs expenses
fig=plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(df['sales'],df['customers'])
ax1.set_title('sales vs customers')
ax1.set_xlabel('sales')
ax1.set_ylabel('customers')


ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(df['sales'], df['customers'], color='r', marker='o')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('sales')
ax2.set_ylabel('customers')


ax3 = fig.add_subplot(2, 2, 3)
#bar_labels = df['month']
bar_values = df['sales']
ax3.bar(bar_labels, bar_values, color='g', alpha=0.5)
ax3.set_title('Bar Plot')
ax3.set_xlabel('sales')
ax3.set_ylabel('customers')
plt.xticks(df['month'],rotation=90)


ax4 = fig.add_subplot(2, 2, 4)
data = df['sales']
ax4.hist(data, bins=4, color='purple', alpha=0.7, edgecolor='black')
ax4.set_title('Histogram')
ax4.set_xlabel('sales')
ax4.set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```
