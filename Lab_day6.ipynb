Parsing XML with Ixml.objectify

```python
from lxml import objectify
path='Performance_MNR (1).xml'
parsed=objectify.parse(open(path))
root=parsed.getroot()
```

```python
root.INDICATOR
```

```python
root
```

```python
data=[]
skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.INDICATOR:
  el_data = {}
  for child in elt.getchildren():
    if child.tag in skip_fields:
      continue
    el_data[child.tag] = child.pyval
  data.append(el_data) 
perf=pd.DataFrame(data)
perf
```

```python
perf.index
```

*HDF5* *FORMAT*

```python
import numpy as np
```

```python
frame2=pd.DataFrame({'a':np.random.randn(100)})
frame2
```

```python
store=pd.HDFStore('mydata.h5')
store['obj1']=frame2
store['obj_col']=frame2['a']
```

```python
store
```

```python
store['obj1']
```

Query operations on HDFStore

```python
store.put('obj2',frame2,format='table')
```

```python
store.select('obj2',where=['index >= 10 and index <15'])
```

```python
store.close()
```

Another way to use these query tools

```python
frame2.to_hdf('mydata.h5','obj3',format='table')
pd.read_hdf('mydata.h5','obj3',where=['index<3'])
```

#interaction with databases

```python
import sqlite3
```

```python
query= """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
c REAL, d INTEGER
);"""
con = sqlite3.connect('mydata.sqlite')
con.execute(query)
```

```python
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
        ('Tallahassee','Florida',2.6, 3),
        ('Sacramento', 'California', 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
```

```python
con.commit()
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall()
rows
```

```python
import pandas as pd
pd.DataFrame(rows,columns=[x[0] for x in cursor.description])
```

```python
import sqlalchemy as sqla
db=sqla.create_engine('sqlite:///mydata.sqlite')
pd.read_sql('select * from test', db)
```
