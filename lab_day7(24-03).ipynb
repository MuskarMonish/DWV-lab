create a dataframe of 7rows and 3columns, fill it with random values 
Impute null values starting from row index 0 to row index 3 and column index 1
also impute null values at column label 2 from row index 0 to 1

```python
import numpy as np
df=pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1]=NA
df.iloc[:2,2]=NA
df
```

delete all rows containing null values in above df

```python
df.dropna()
```

delete a row only if it doesnt contain a min of 2 non missing values

```python
df.dropna(thresh=2)
```

fill all null values with 0

```python
df.fillna(0)
```

fill all missing values at column index 1 with 50 and column index 2 with 100

```python
df.fillna({1:50,2:100})
```

fill all null values with forward fill


```python
data.fillna(method='ffill')
```

fill all null values with mean value of the column values

```python
df.fillna(df.mean())
```

create df containing 2 columns name and age. transform age column  using a map method and user defined fn to create a new column age_in_months

```python
data2 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],'age': [25, 30, 35, 40]}
df3 = pd.DataFrame(data2)
df3
```

```python
def age_in_months(age):
  return age * 12
# use map() to apply the age_in_months function to the age column
df3['age_in_months'] = df3['age'].map(age_in_months)
# print the resulting dataframe
print(df3)
```
